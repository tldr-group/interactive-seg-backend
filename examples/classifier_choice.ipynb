The goal of this notebook is simply to demonstrate how easy it is to swap out parts of the workflow by changing the classifier

```python
from interactive_seg_backend.file_handling import load_image, load_labels
from interactive_seg_backend.configs.types import ClassifierNames
from interactive_seg_backend import FeatureConfig, TrainingConfig, featurise, train_and_apply
```

```python
image = load_image('../tests/data/0.tif')
labels = load_labels('../tests/data/0_labels.tif')

feat_cfg = FeatureConfig()
train_cfg = TrainingConfig(feature_config=feat_cfg)
```

```python
feats = featurise(image, train_cfg)
```

```python
classifier_names: tuple[ClassifierNames, ...] = ('linear_regression', 'logistic_regression', 'random_forest', 'xgb')

preds = {}
for classifier in classifier_names:
    print(f"Training and applying {classifier}...")
    train_cfg.classifier = classifier
    pred, _, _ = train_and_apply(feats, labels, train_cfg)
    preds[classifier] = pred
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, len(classifier_names) + 1, figsize=(15, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

for i, classifier in enumerate(classifier_names):
    axes[i + 1].imshow(preds[classifier], cmap='jet', interpolation=None)
    axes[i + 1].set_title(f'{classifier}')
    axes[i + 1].axis('off')
```
